```python
csv_path_1 = "/content/drive/MyDrive/ds_Rathod Srushtitha/csv_files/historical_data.csv"
df1 = pd.read_csv(csv_path_1)

csv_path_2 = "/content/drive/MyDrive/ds_Rathod Srushtitha/csv_files/fear_greed_index.csv"
df2 = pd.read_csv(csv_path_2)
```

```python
# converting timestamp IST column in df1 into datetime
df1['Date'] = pd.to_datetime(df1['Timestamp IST'], format = "%d-%m-%Y %H:%M")
df1.head()
```

```python
# converting date column in df2 into datetime
df2['Date'] = pd.to_datetime(df2['date'])
df2.head()
```

```python
df2.drop('date', axis = 1, inplace = True)
df2.head()
```

```python
# Merging both dataframes df1 and df2 on Date column
merged_data = pd.merge(df1, df2, on = 'Date', how = 'inner')
merged_data.head()
```

```python
merged_data.shape
```

```python
# save merged_data
merged_data.to_csv("/content/drive/MyDrive/ds_Rathod Srushtitha/csv_files/merged_data.csv", index = False)
```

```python
#  Profitability flag
merged_data['Profitable'] = merged_data['Closed PnL'] > 0
```

```python
# PnL per dollar traded (normalized metric)
merged_data['PnL_per_USD'] = merged_data['Closed PnL'] / merged_data['Size USD']
merged_data['PnL_per_USD'] = merged_data['PnL_per_USD'].replace([float('inf'), -float('inf')], 0).fillna(0)
```

```python
# Sentiment level: convert classification to levels
def sentiment_level(classification):
    if classification.lower() in ['extreme fear', 'fear']:
        return 'Negative'
    elif classification.lower() in ['greed', 'extreme greed']:
        return 'Positive'
    else:
        return 'Neutral'
merged_data['Sentiment_Level'] = merged_data['classification'].apply(sentiment_level)
```

```python
# Day of week
merged_data['Day_of_Week'] = merged_data['Date'].dt.day_name()
```

```python
#  Volume category
merged_data['Volume_Category'] = pd.qcut(merged_data['Size USD'], q=3, labels=['Low', 'Medium', 'High'])
```

```python
merged_data.head()
```

```python
# Save to Drive
merged_data.to_csv("/content/drive/MyDrive/ds_Rathod Srushtitha/csv_files/cleaned_merged_data.csv", index=False)
```

Sentiment vs Profitability Rate

Are trades more profitable during Greed vs Fear?

```python
# Sentiment vs % of profitable trades
sentiment_profit = merged_data.groupby('Sentiment_Level')['Profitable'].mean().reset_index()

# Bar plot
plt.figure(figsize=(6,4))
sns.barplot(data=sentiment_profit, x='Sentiment_Level', y='Profitable', palette='viridis')
plt.title('Profitability Rate by Sentiment Level')
plt.ylabel('Profitable')
plt.xlabel('Sentiment Level')
plt.ylim(0, 1)
plt.grid(True, axis='y')
plt.show()
```

Day of Week Analysis

Which weekday has better profitability?

```python
# Day vs Profitability
day_profit = merged_data.groupby('Day_of_Week')['Profitable'].mean().reset_index()

# Order the weekdays
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_profit['Day_of_Week'] = pd.Categorical(day_profit['Day_of_Week'], categories=days_order, ordered=True)
day_profit = day_profit.sort_values('Day_of_Week')

plt.figure(figsize=(8,4))
sns.barplot(data=day_profit, x='Day_of_Week', y='Profitable', palette='coolwarm')
plt.title('Profitability by Day of Week')
plt.ylabel('Proportion of Profitable Trades')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```

 Heatmap: Sentiment vs Volume vs Profitability

Are high-volume trades more profitable in certain sentiment zones?

```python
# Create a pivot table
pivot = merged_data.pivot_table(values='Profitable',
                                 index='Sentiment_Level',
                                 columns='Volume_Category',
                                 aggfunc='mean')

# Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(pivot, annot=True, fmt=".2f", cmap='YlGnBu')
plt.title('Profitability by Sentiment Level and Volume Category')
plt.ylabel('Sentiment Level')
plt.xlabel('Volume Category')
plt.show()
```
